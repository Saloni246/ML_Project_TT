```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# READING DATASET

```python
df=pd.read_csv('time_series_2019-ncov-Confirmed.csv')
df
```

```python
df.dropna(axis=1)
```

### SELECTING COLUMNS CONTAINING ONLY DATES

```python
df1=df.iloc[:,4:65]
df1
```

```python
#printing percentage of each column depending on 0 confirmed cases

a=print (df1.isin([0]).mean())
```

```python
#Dropping all the columns or dates having 0 confirmed cases above 70% of the total

df2 = df1.loc[:, df1.isin([0]).mean() < .70]       #named this new dataset as df2
df2
```

# COUNTRY-WISE PLOT

```python
labels=df.iloc[:,1] #selcting country column as label from original dataset
y=df.iloc[:,-1]     #selecting total number of cases from last column
plt.figure(figsize=(30,10))

#Drawing Pie-Chart
patches,text=plt.pie(y)
plt.title('CONFIRMED CASES COUNTRY-WISE',size=24)
plt.legend(patches,labels,title='COUNTRIES',loc='centre left',prop={'size':10},bbox_to_anchor=(-0.1, 1.))
```

# DATE-WISE PLOT

```python
#finding sum of each column from new dataset df2
df2.sum(axis=0)
```

```python
dates=df2.keys()
total_cases_on_date=[]
for i in dates[:]:
    sum_of_cases=df2[i].sum()
    total_cases_on_date.append(sum_of_cases)
    print(f'{i}: {sum_of_cases}')
```

```python
#Drawing Pie-Chart

plt.figure(figsize=(30,10))

plt.pie(total_cases_on_date,labels=dates,autopct='%1.0f%%', pctdistance=0.75,labeldistance=1.05)
plt.title('CONFIRMED CASES DATE-WISE',size=24)
plt.legend(dates,title='DATES',loc='centre left',prop={'size':10},bbox_to_anchor=(-0.1, 1.))
```

# CONTINENT-WISE PLOT

```python
cont=pd.read_csv('c2c.csv') #reading continent and country dataset
cont
```

```python
#Making dictionary of country name and total cases in each country
country_names=dict()

for i in range(487):
    country_names[df.iloc[i][1]]=country_names.get(df.iloc[i][1],0)+df.iloc[i][-1]   
```

```python
#Making dictionary of continent name and total cases in each continent
continent_names=dict()

for i in country_names:
    names=cont.loc[cont['Country']==i]
    if len(names)>0:
        continent_names[names.iloc[0][0]]=continent_names.get(names.iloc[0][0],0)+country_names[i]
        
continent_names      #printing total cases in each continent  
```

```python
#Drawing Pie-Chart

plt.figure(figsize=(30,10))
x=continent_names.values()
label=continent_names.keys()
explode=(0,0,0.1,0,0.3,0)

plt.pie(x,labels=label,autopct='%1.0f%%', pctdistance=0.75,labeldistance=1.09,explode=explode)
plt.title('CONFIRMED CASES CONTINENT-WISE',size=24)
plt.legend(label,title='CONTINENTS',loc='centre left',prop={'size':10},bbox_to_anchor=(-0.1, 1.))
```
